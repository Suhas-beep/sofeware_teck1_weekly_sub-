In this task, I worked on normalizing the MNIST dataset. First, I loaded the dataset using Keras and stored it into training and testing sets.

Then, I reshaped the dataset to include a single channel. Since the images are grayscale, they only have one channel, so I reshaped them into the format (height, width, 1).

After that, I checked the pixel values of the images by printing the minimum and maximum values. I observed that the values range from 0 to 255.

Next, I used ImageDataGenerator with rescale=1.0/255.0 to normalize the dataset. This means all pixel values are scaled between 0 and 1, which makes the data easier to work with in machine learning.

I then created iterators for both training and testing data using the generator. These iterators help in processing the data in batches.

Finally, I took one batch from the training data and printed its shape and pixel range to confirm that the normalization worked correctly. The output showed that the values arenow between 0 and 1.

```python
from keras.datasets import mnist
from tensorflow.keras.preprocessing.image import ImageDataGenerator
# load dataset
(trainX, trainy), (testX, testy) = mnist.load_data()
# reshape dataset to have a single channel
width, height, channels = trainX.shape[1], trainX.shape[2], 1
trainX = trainX.reshape((trainX.shape[0], width, height, channels))
testX = testX.reshape((testX.shape[0], width, height, channels))
# report per-image mean
print('Means train=%.3f, test=%.3f' % (trainX.mean(), testX.mean()))
# create generator that centers pixel values
datagen = ImageDataGenerator(featurewise_center=True)
# calculate the mean on the training dataset
datagen.fit(trainX)
print('Data Generator Mean: %.3f' % datagen.mean.item())
# demonstrate effect on a single batch of samples
iterator = datagen.flow(trainX, trainy, batch_size=64)
# get a batch
batchX, batchy = next(iterator)
# mean pixel value in the batch
print(batchX.shape, batchX.mean())
# demonstrate effect on entire training dataset
iterator = datagen.flow(trainX, trainy, batch_size=len(trainX), shuffle=False)
# get a batch
batchX, batchy = next(iterator)
# mean pixel value in the batch
print(batchX.shape, batchX.mean())
```

In this task, I worked on centering the MNIST dataset. First, I loaded the dataset and reshaped it to include a single channel since the images are grayscale.

After that, I printed the mean pixel values of both the training and testing datasets to understand their average intensity.

Then, I used ImageDataGenerator with the parameter featurewise_center=True. This setting subtracts the mean value of the dataset from each pixel, which helps to center the data around zero.

I used datagen.fit(trainX) to calculate the mean value from the training dataset. This step is important because the generator needs to know the average pixel value before applying the centering process.

Next, I created an iterator to process the data in batches and retrieved one batch using next(). I printed the shape and mean of the batch to confirm that the pixel values are now centered around zero.

Finally, I applied the same process to the entire training dataset and confirmed that the mean is close to zero after centering.

Overall, this task helped me understand how centering works and why it is useful for improving the performance of machine learning models.

```python
from keras.datasets import mnist
from tensorflow.keras.preprocessing.image import ImageDataGenerator
# load dataset
(trainX, trainy), (testX, testy) = mnist.load_data()
# reshape dataset to have a single channel
width, height, channels = trainX.shape[1], trainX.shape[2], 1
trainX = trainX.reshape((trainX.shape[0], width, height, channels))
testX = testX.reshape((testX.shape[0], width, height, channels))
# report pixel means and standard deviations
print('Statistics train=%.3f (%.3f), test=%.3f (%.3f)' % (trainX.mean(), trainX.std(), testX.mean(), testX.std()))
# create generator that centers pixel values
datagen = ImageDataGenerator(featurewise_center=True, featurewise_std_normalization=True)
# calculate the mean on the training dataset
datagen.fit(trainX)
print('Data Generator mean=%.3f, std=%.3f' % (datagen.mean.item(), datagen.std.item()))
# demonstrate effect on a single batch of samples
iterator = datagen.flow(trainX, trainy, batch_size=64)
# get a batch
batchX, batchy = next(iterator)
# pixel stats in the batch
print(batchX.shape, batchX.mean(), batchX.std())
# demonstrate effect on entire training dataset
iterator = datagen.flow(trainX, trainy, batch_size=len(trainX), shuffle=False)
# get a batch
batchX, batchy = next(iterator)
# pixel stats in the batch
print(batchX.shape, batchX.mean(), batchX.std())
```

In this task, I worked on standardizing the MNIST dataset. First, I loaded the dataset and reshaped it to include a single channel since the images are grayscale.

Then, I printed the mean and standard deviation of the dataset to understand how the pixel values are distributed.

After that, I used ImageDataGenerator with both featurewise_center=True and featurewise_std_normalization=True. This means the data is first centered by subtracting the mean and then scaled by dividing by the standard deviation.

I used datagen.fit(trainX) to calculate the mean and standard deviation from the training dataset. These values are then used to standardize the data.

Next, I created an iterator and retrieved a batch of data to check the results. I printed the shape, mean, and standard deviation of the batch. From the output, I observed that the mean is close to 0 and the standard deviation is close to 1.

Finally, I applied the same process to the entire dataset and confirmed that the values are properly standardized.

Overall, this task helped me understand how standardization works by scaling the data so that it has a mean of 0 and a standard deviation of 1, which is useful for improving model performance.
